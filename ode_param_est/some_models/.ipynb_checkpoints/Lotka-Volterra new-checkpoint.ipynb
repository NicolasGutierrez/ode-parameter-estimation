# Parameter estimation techniques in Lotka-Volterra model

In this document, different parameter estimation techniques for nonlinear ODE models are applied to a **Lotka-Volterra model** predator-prey model. First, we will use _nonlinear least squares_ to find the best fit. This method requires an initial estimation for the value of the parameters, and an important downside is that the results are not robust under different initial estimations, since the cost function to be minimized in order to find the best fit could have local minima to which the solution tends instead of the global minimum. We will work our way around this, seizing the fact that the Lotka-Volterra model is linear in its parameters and use a _least squares regression_ on the parameters. Nevertheless, we will see that the obtained estimates are very sensitive to noise, so we will use nonlinear least squares, with the previously obtained parameters as initial estimates, to improve our estimation. 

First of all, we define some necessary functions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
from scipy import integrate
from scipy import optimize
from scipy import interpolate
plt.style.use('ggplot')

def lv_equation(t, x, *args):
    """
    Returns the Lotka-Volterra differential equation for two populations.
    Should receive six parameters (r1, r2, b11, b12, b21, b22)
    """
    r1 = args[0]
    b12 = args[1]
    b13 = args[2]
    b14 = args[3]
    r2 = args[4]
    b21 = args[5]
    b23 = args[6]
    b24 = args[7]
    r3 = args[8]
    b31 = args[9]
    b32 = args[10]
    b34 = args[11]
    r4 = args[12]
    b41 = args[13]
    b42 = args[14]
    b43 = args[15]
    r = np.array([r1,r2,r3,r4])
    b = np.array([[0,b12,b13,b14], [b21,0,b23,b24], [b31,b32,0,b34], [b41,b42,b43,0]])
    return x * (r + b @ x)  # x_i = x_i (r_i + \sum b_ij x_j)

def add_noise(data, sigma=1):
    """
    Implements additive Gaussian noise with mean 0 and standard deviation sigma
    to each entry of the given data
    """
    return data + np.random.normal(scale=sigma, size=np.shape(data))

def mult_noise(data, sigma=1):
    """
    Implements multipliactive Gaussian noise with mean 1 and standard deviation sigma
    to each entry of the given data
    """
    return data * np.random.normal(scale=sigma, size=np.shape(data))

def clean_data(data, eps):
    """
    Reduces to zero the entries of data with value < eps
    """
    data[data < eps] = 0
    return data
```

## Data generation

First, the necessary parameters are defined. Here, $x_0$ is an array containing the initial population sizes, while $t_0$ and $t_f$ define the interval of the ODE to be integrated. Finally, we generate $t_{points}$, an array containing the time points in which the solution of the ODE is to be evaluated.

Next, we generate noisy data. For this purpose, we first generate data with the previous parameters and add Gaussian noise with mean $0$ and standard deviation $\sigma$. Then, we simply set the values below a certain threshold $\epsilon$ to zero. We show plots of the original data and the noisy data side to side.

```python
# ODE Parameters
r1 = 2
b12 = -0.6
b13 = 0
b14 = -0.2
r2 = 1
b21 = 0.7
b21 = 0.6
b23 = -0.6
b24 = -0.2
r3 = 0
b31 = 0
b32 = 0.6
b34 = -0.2
r4 = -3
b41 = 0.2
b42 = 0.2
b43 = 0.2

params = (r1,b12,b13,b14,r2,b21,b23,b24,r3,b31,b32,b34,r4,b41,b42,b43)

# Additional conditions
t0 = 0     
tf = 10
x0 = np.array([5,4,3,2])

# Time points to be displayed
num_tpoints = 200
t_points = np.linspace(t0, tf, num_tpoints)

# Noise parameters
sigma1 = 0.05
sigma2 = 0.15
sigma3 = 0.5

# Data below this value is considered to be zero
eps = 1E-6

# Data generation and stuff
sol_lv = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=params, t_eval=t_points)
data_lv1 = add_noise(sol_lv.y, sigma1)
data_lv1 = clean_data(data_lv1, eps)

data_lv2 = add_noise(sol_lv.y, sigma2)
data_lv2 = clean_data(data_lv2, eps)

data_lv3 = add_noise(sol_lv.y, sigma3)
data_lv3 = clean_data(data_lv3, eps)

# Plotting
fig, (ax0,ax1,ax2,ax3) = plt.subplots(4, 1, figsize=(8,15))
ax0.plot(t_points, sol_lv.y[0], 'r*', label='x1')
ax0.plot(t_points, sol_lv.y[1], 'g*', label='x2') 
ax0.plot(t_points, sol_lv.y[2], 'b*', label='x3') 
ax0.plot(t_points, sol_lv.y[3], 'm*', label='x4') 
ax0.set_title('sigma = 0')
ax0.legend(loc='best')

ax1.plot(t_points, data_lv1[0], 'r*', label='x1')
ax1.plot(t_points, data_lv1[1], 'g*', label='x2') 
ax1.plot(t_points, data_lv1[2], 'b*', label='x3') 
ax1.plot(t_points, data_lv1[3], 'm*', label='x4') 
ax1.set_title('sigma = %.2f' % sigma1)

ax2.plot(t_points, data_lv2[0], 'r*', label='x1')
ax2.plot(t_points, data_lv2[1], 'g*', label='x2') 
ax2.plot(t_points, data_lv2[2], 'b*', label='x3') 
ax2.plot(t_points, data_lv2[3], 'm*', label='x4') 
ax2.set_title('sigma = %.2f' % sigma2)

ax3.plot(t_points, data_lv3[0], 'r*', label='x1')
ax3.plot(t_points, data_lv3[1], 'g*', label='x2') 
ax3.plot(t_points, data_lv3[2], 'b*', label='x3') 
ax3.plot(t_points, data_lv3[3], 'm*', label='x4') 
ax3.set_title('sigma = %.2f' % sigma3)
ax3.set_xlabel('time')
plt.show()
```

## Parameter estimation

### Linear least squares

We follow the treatment in Stein et al. (2013) for our linear least squares procedure. To do this, first we define the appropriate matrices and vectors so that our differential equation is written in matrix equation form $A x = b$. To do this, we first discretize the differential equation. Our $b$ will be a matrix of 'log differentials' (roughly, the left side of the original differential equation), $x$ is a matrix containing all the ODE parameters.

Since scipy methods are optimized to solve these systems in the case x and b are one-dimensional arrays, we separate the equation in two parts,

\begin{align}
&A_1 x_1 = b_1 \\
&A_2 x_2 = b_2, 
\end{align}

```python
def into_matrix_form(data, num_tpoints):
    b1 = np.zeros(num_tpoints)
    b2 = np.zeros(num_tpoints)
    A1 = np.zeros((num_tpoints, 2))
    A2 = np.zeros((num_tpoints, 2))
    A1[:,0] = np.ones(num_tpoints)    # sets first column equal to ones
    A2[:,0] = np.ones(num_tpoints)    # ditto
    for i in range(num_tpoints-1):
        if data[0,i+1]==0 or data[0,i]==0:
            b1[i] = 0
        else:
            b1[i] = np.log(data[0,i+1]/data[0,i])/(t_points[i+1]-t_points[i])
        if data[1,i+1]==0 or data[1,i]==0:
            b2[i] = 0
        else:
            b2[i] = np.log(data[1,i+1]/data[1,i])/(t_points[i+1]-t_points[i])
        A1[i,1] = data[1,i]
        A2[i,1] = data[0,i]
    return (A1,b1,A2,b2)

def obtain_params(data, num_tpoints):
    A1,b1,A2,b2 = into_matrix_form(data, num_tpoints)
    reg1 = optimize.lsq_linear(A1, b1)    # returns (r1, b12)
    reg2 = optimize.lsq_linear(A2, b2)    # returns (r2, b21)
    return tuple(reg1.x) + tuple(reg2.x)
    
ls_fitted_params1 = obtain_params(data_lv1, num_tpoints)
ls_fitted_params2 = obtain_params(data_lv2, num_tpoints)
ls_fitted_params3 = obtain_params(data_lv3, num_tpoints)

print(ls_fitted_params1)

"""
ls_fitted_data1 = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=ls_fitted_params1, t_eval=t_points)
ls_fitted_data2 = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=ls_fitted_params2, t_eval=t_points)
ls_fitted_data3 = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=ls_fitted_params3, t_eval=t_points)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10,10))
ax1.plot(t_points, ls_fitted_data1.y[0], 'r-', label='fit')
ax1.plot(t_points, ls_fitted_data1.y[1], 'b-', label='fit')
ax1.plot(t_points, data_lv1[0], 'm*', label='data')
ax1.plot(t_points, data_lv1[1], 'g*', label='data')
ax1.set_title(f'sigma = {sigma1}')
ax1.legend(loc='best')
ax1.set_ylabel('Population')

ax2.plot(t_points, ls_fitted_data2.y[0], 'r-', label='fit')
ax2.plot(t_points, ls_fitted_data2.y[1], 'b-', label='fit')
ax2.plot(t_points, data_lv2[0], 'm*', label='data')
ax2.plot(t_points, data_lv2[1], 'g*', label='data')
ax2.set_title(f'sigma = {sigma2}')
ax2.set_ylabel('Population')

ax3.plot(t_points, ls_fitted_data3.y[0], 'r-', label='fit')
ax3.plot(t_points, ls_fitted_data3.y[1], 'b-', label='fit')
ax3.plot(t_points, data_lv3[0], 'm*', label='data')
ax3.plot(t_points, data_lv3[1], 'g*', label='data')
ax3.set_title(f'sigma = {sigma3}')
ax3.set_ylabel('Population')
ax3.set_xlabel('Time')

plt.show()
"""
```

```python
# SUMMARY OF RESULTS
print("For sigma = %.2f, the errors in each parameter were: \n" % 
      sigma1, np.abs(np.array(ls_fitted_params1) - np.array(params)))
print("\nFor sigma = %.2f, the errors in each parameter were: \n" % 
      sigma2, np.abs(np.array(ls_fitted_params2) - np.array(params)))
print("\nFor sigma = %.2f, the errors in each parameter were: \n" % 
      sigma3, np.abs(np.array(ls_fitted_params3) - np.array(params)))
```

### Nonlinear Least Squares

We now use our previously obtained parameter estimates as initial estimates for a nonlinear least squares optimization. A comparison will be made to establish how much our estimates are improved by carrying out this procedure.

```python
def lv_residuals1(params):
    sim_data = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=params, t_eval=t_points)
    return sim_data.y[0] - data_lv1[0] + sim_data.y[1] - data_lv1[1]

def lv_residuals2(params):
    sim_data = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=params, t_eval=t_points)
    return sim_data.y[0] - data_lv2[0] + sim_data.y[1] - data_lv2[1]

def lv_residuals3(params):
    sim_data = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=params, t_eval=t_points)
    return sim_data.y[0] - data_lv3[0] + sim_data.y[1] - data_lv3[1]

nls_fitted_params1 = optimize.least_squares(lv_residuals1, ls_fitted_params1)
nls_fitted_params2 = optimize.least_squares(lv_residuals2, ls_fitted_params2)
nls_fitted_params3 = optimize.least_squares(lv_residuals3, ls_fitted_params3)

initial_guess1 = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=ls_fitted_params1, t_eval=t_points)
nls_fitted_data1 = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=nls_fitted_params1.x, t_eval=t_points)

initial_guess2 = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=ls_fitted_params2, t_eval=t_points)
nls_fitted_data2 = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=nls_fitted_params2.x, t_eval=t_points)

initial_guess3 = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=ls_fitted_params3, t_eval=t_points)
nls_fitted_data3 = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=nls_fitted_params3.x, t_eval=t_points)

#plt.plot(t_points, initial_guess.y[0], 'b-', label='Initial Guess')
#plt.plot(t_points, initial_guess.y[1], 'b-', label='Initial Guess')
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10,10))
ax1.plot(t_points, nls_fitted_data1.y[0], 'r-', label='Fitted Parameters')
ax1.plot(t_points, nls_fitted_data1.y[1], 'b-', label='Fitted Parameters')
ax1.plot(t_points, data_lv1[0], 'm*', label='Data')
ax1.plot(t_points, data_lv1[1], 'g*', label='Data')
ax1.set_title('sigma = %.2f' % sigma1)
ax1.legend(loc='upper left')
ax1.set_ylabel('Population')

ax2.plot(t_points, nls_fitted_data2.y[0], 'r-', label='Fitted Parameters')
ax2.plot(t_points, nls_fitted_data2.y[1], 'b-', label='Fitted Parameters')
ax2.plot(t_points, data_lv2[0], 'm*', label='Data')
ax2.plot(t_points, data_lv2[1], 'g*', label='Data')
ax2.set_title('sigma = %.2f' % sigma2)
ax2.legend(loc='upper left')
ax2.set_ylabel('Population')

ax3.plot(t_points, nls_fitted_data2.y[0], 'r-', label='Fitted Parameters')
ax3.plot(t_points, nls_fitted_data2.y[1], 'b-', label='Fitted Parameters')
ax3.plot(t_points, data_lv2[0], 'm*', label='Data')
ax3.plot(t_points, data_lv2[1], 'g*', label='Data')
ax3.set_title('sigma = %.2f' % sigma3)
ax3.legend(loc='upper left')
ax3.set_ylabel('Population')
ax3.set_xlabel('Time')

plt.show()
```

```python
# SUMMARY OF RESULTS
print("\nFor sigma = %.2f:" % sigma1)
print("The errors in each parameter in LLS were : \n", np.abs(np.array(ls_fitted_params1) - np.array(params)))
print("The errors in each parameter in NLS were : \n", np.abs(np.array(nls_fitted_params1.x) - np.array(params)))

print("\nFor sigma = %.2f:" % sigma2)
print("The errors in each parameter in LLS were : \n", np.abs(np.array(ls_fitted_params2) - np.array(params)))
print("The errors in each parameter in NLS were : \n", np.abs(np.array(nls_fitted_params2.x) - np.array(params)))

print("\nFor sigma = %.2f:" % sigma3)
print("The errors in each parameter in LLS were : \n", np.abs(np.array(ls_fitted_params3) - np.array(params)))
print("The errors in each parameter in NLS were : \n", np.abs(np.array(nls_fitted_params3.x) - np.array(params)))
```
